```python
import pandas as pd

def download_csv_from_url(repo_url, file_name):
    """
    Descargo un archivo CSV desde la URL proporcionada, lo guardo local, retorno un df.
   Por problemas con detectar el delimitador treato con un for.
    """
    separador = [',', ';', '\t']
    for sep in separador:
        try:
            df = pd.read_csv(repo_url, sep=sep)
            df.to_csv(file_name, index=False)
            print(f"Archivo descargado y guardado como '{file_name}'")
            #f-string es un tipo de string en el cual le puedo meter una variable en medio
            return df
        except Exception as e:
            continue
    print("Error al leer el CSV: No se pudo detectar el formato correcto.")
    return None

# URLs proporcionadas
url_disney = "https://desafio-rkd.s3.amazonaws.com/disney_plus_titles.csv"
url_netflix = "https://desafio-rkd.s3.amazonaws.com/netflix_titles.csv"

# Descargo los archivos
df_disney = download_csv_from_url(url_disney, "disney_plus_titles.csv")
df_netflix = download_csv_from_url(url_netflix, "netflix_titles.csv")

# Verifico si se descargo
if df_disney is not None:
    print("Disney+ CSV preview:")
    display(df_disney.head())

if df_netflix is not None:
    print("Netflix CSV preview:")
    display(df_netflix.head())

```
